```python
import csv
import random
def loadDataset(filename, split, trainingSet=[], trainingSetLabels=[] , testSet=[], testSetLabels=[]):
    with open(filename, 'r') as csvfile:
        lines = csv.reader(csvfile)
        dataset = list(lines)
        for x in range(len(dataset)-1):
            for y in range(4):
                dataset[x][y] = float(dataset[x][y])
                if random.random() < split:
                    trainingSet.append(dataset[x][0:4])
                    trainingSetLabels.append(dataset[x][4])
                else:
                    testSet.append(dataset[x][0:4])
                    testSetLabels.append(dataset[x][4])
trainingSet=[]
testSet=[]
trainingSetLabels=[]
testSetLabels=[]
loadDataset('iris.csv', 0.5, trainingSet, trainingSetLabels, testSet, testSetLabels)
trainingSetLabels
```

```python
print ('Train: ' + repr(len(trainingSet)))
print ('Test: ' + repr(len(testSet)))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

clf = SVC(kernel='linear')
clf.fit(trainingSet,trainingSetLabels)
y_pred = clf.predict(testSet)
print(accuracy_score(testSetLabels,y_pred))
```
