```python
def plot_confusion_matrix(cm, imageName, labels = []):
    sns.set(rc={'figure.facecolor':'white'})
    sns.set(rc={'figure.figsize':[20, 9]})
    cm_norm = normaliser(cm) # normalize confusion matrix by summing the rows, and diviing each value by it, repressented as a percentage
    cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["white","blue"])
    ax = sns.heatmap(cm_norm, vmin=0, vmax=100,
                 linecolor='black', linewidth=1,
                     annot=True, fmt='.2f', xticklabels=labels, yticklabels=labels, cmap=cmap)
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    for i, t in enumerate(ax.texts):
        t.set_text('{0} ({1:.2f}%)'.format(cm.flatten()[i], cm_norm.flatten()[i]))
    
    #plt.show()
    fig = ax.get_figure()
    fig.savefig(imageName)
```
